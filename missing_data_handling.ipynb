```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df
```

```python
var_14_values = np.array(df['VAR_14'])
var_14_values = var_14_values[np.where(var_14_values != '?')]
var_14_values = np.array(var_14_values, dtype=np.float64)
```

```python
plt.boxplot(var_14_values) # No outliers, we can use mean to handle missing values
```

```python
# Replace VAR_14 with mean
df['VAR_14'] = pd.to_numeric(df['VAR_14'], errors='coerce').fillna(var_14_values.mean())
```

```python
var_6_values = np.array(df['VAR_6'])
var_6_mode = stats.mode(var_6_values)[0][0]
var_6_values[np.where(var_6_values == '?')] = var_6_mode
```

```python
df['VAR_6'] = var_6_values
```
